```python
%matplotlib notebook
import pandas as pd
import itertools
from mpl_toolkits.mplot3d import Axes3D
import mpl_toolkits.mplot3d as m3d
import matplotlib.pyplot as plt
import numpy as np
```

```python
r = range(95, 256)
g = range(40, 256)
b = range(20, 256)

rgb = [r, g, b]

comb = list(itertools.product(*rgb))
li = []
for i in comb:
    if (i[0] > i[1]) & (i[0] > i[2]) & (i[0] - i[1] > 15):
        li.append(i)
print len(li)
```

```python
print li[0:5]
hex = []
for i in li:
    hex.append('#%02x%02x%02x' % i)
```

```python
print hex[0:5]
```

```python
allSkinTones = pd.DataFrame(li)
allSkinTones['HexCodes'] = hex
allSkinTones = allSkinTones.rename(index=str, columns={0: "R", 1: "G", 2: "B"})
```

```python
allSkinTones.info()
```

```python
R = allSkinTones['R'].astype('float64')
G = allSkinTones['G'].astype('float64')
B = allSkinTones['B'].astype('float64')

# R = list(allSkinTones['R'])
# G = list(allSkinTones['G'])
# B = list(allSkinTones['B'])
color = allSkinTones['HexCodes']

ax = m3d.Axes3D(plt.figure(figsize = (8, 7)))
ax.scatter3D(R, G, B, c = color, depthshade=False)
ax.set_xlim3d(0,255)
ax.set_ylim3d(0,255)
ax.set_zlim3d(0,255)
ax.set_xlabel('R', fontsize = 20)
ax.set_ylabel('G', fontsize = 20)
ax.set_zlabel('B', fontsize = 20)
plt.show()
```
