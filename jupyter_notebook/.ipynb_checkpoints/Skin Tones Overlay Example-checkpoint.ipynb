```python
from mpl_toolkits.mplot3d import Axes3D
import mpl_toolkits.mplot3d as m3d
import matplotlib.pyplot as plt

jet=plt.get_cmap('jet')

x = [1,2,3,4]
y = [1,2,3,4]
z = [1,1,1,1]

a = [2,3,4,5]
b = [1,2,3,4]
c = [2,2,2,2]

d = [3,4,5,6]
e = [1,2,3,4]
f = [3,3,3,3]

ax = m3d.Axes3D(plt.figure())
ax.scatter3D(x, y, z, s=100, c=z, vmin=1, vmax=5, cmap=jet)
ax.scatter3D(a, b, c, s=100, c=c, vmin=1, vmax=5, cmap=jet)
ax.scatter3D(d, e, f, s=100, c=f, vmin=1, vmax=5, cmap=jet, alpha=0.5)

plt.show()
```
